# Oversampling

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# 가중치를 만듭니다.
weights = {0: .1, 1: 0.9}

# 가중치를 부여한 랜덤 포레스트 분류기를 만듭니다.
RandomForestClassifier(class_weight=weights)

# 균형잡힌 클래스 가중치로 랜덤 포레스트 모델을 훈련합니다.
RandomForestClassifier(class_weight="balanced")
```

```python
df_X = data.drop('depvar', axis=1)
df_y = data['depvar']
```

```python
i_class0 = np.where(df_y == 0)[0]
i_class1 = np.where(df_y == 1)[0]
n_class0 = len(i_class0)
n_class1 = len(i_class1)
```

```python
df_y = df_y.to_numpy()
```

```python
# 클래스 0의 샘플 개수만큼 클래스 1에서 중복을 허용하여 랜덤하게 샘플을 선택합니다.
i_class1_upsampled = np.random.choice(i_class1, size=n_class0, replace=True)

# 클래스 1의 업샘플링된 타깃 벡터와 클래스 0의 타깃 벡터를 합칩니다.
df_y_up=np.concatenate((df_y[i_class1_upsampled], df_y[i_class0]))
```

```python
df_X = df_X.to_numpy()
```

```python
# 클래스 1의 업샘플링된 특성 행렬과 클래스 0의 특성 행렬을 합칩니다.
df_X_up=pd.DataFrame(np.vstack((df_X[i_class1_upsampled,:], df_X[i_class0,:])))
```

```python
# upsampling 후 data
i_class0_up = np.where(df_y_up == 0)[0]
i_class1_up = np.where(df_y_up == 1)[0]
n_class0 = len(i_class0_up)
n_class1 = len(i_class1_up)
print(n_class0)
print(n_class1)
```

```python
df_X_col = data.drop('depvar', axis=1)
df_X_up = pd.DataFrame(df_X_up)
df_X_up.columns = list(df_X_col.columns)
```

```python
df_y_up = pd.DataFrame(df_y_up)
df_y_up.columns=['depvar']
```

```python
df = pd.concat([df_X_up, df_y_up], axis=1)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score, recall_score, precision_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
def target_profit(model, threshold, X_train, y_train, X_test, y_test, profit):
    
    model.fit(X_train, y_train)
    proba = model.predict_proba(X_test)
    
    predict = proba[:,1].copy()
    predict[predict > threshold] = 1
    predict[predict <= threshold] = 0

    target = pd.DataFrame(predict)
    model_profit = pd.DataFrame(profit).reset_index()
    model_profit = model_profit.drop('index', axis=1)
    target_profit = pd.concat([target, model_profit], axis=1)
    target_profit = target_profit.rename(columns = { 0 : 'result'})

    actual_profit = profit.sum()
    
    predict_profit = 0
    for i in range(len(target_profit)):
        if target_profit['result'][i] == 0:
            predict_profit = predict_profit + target_profit['profit'][i]
        else:
            pass
            
    profit_diff = predict_profit - actual_profit
    
    
    return threshold, profit_diff
```

```python
def plot_optimal_threshold(dataframe, name):
    
    plt.figure(figsize=(12, 9))
    plt.plot(dataframe['threshold'], dataframe['result'], marker = 'o', label = name)
    plt.title('Profit - Threshold', fontsize = 16)
    plt.legend(loc = 'best')
    plt.xticks([0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95])
    plt.xlabel('Threshold')
    plt.ylabel('Profit')
    plt.grid()
    plt.show()
```

```python
def cut_off(y, threshold):
    
    Y=y[:,1].copy()
    Y[Y > threshold] = 1
    Y[Y <= threshold] = 0
    return Y


def threshold_score(model, X_test, y_test, thresholds):
    
    proba_y = model.predict_proba(X_test)
    Y_pred = cut_off(proba_y, thresholds)
    
    score = metrics.accuracy_score(y_test, Y_pred)
    cm = confusion_matrix(y_test, Y_pred)
    precision = precision_score(y_test, Y_pred)
    recall = recall_score(y_test, Y_pred)  
    
    print('confusion_martix','\n', cm)
    print('Accuracy: ', score*100, '%')
    print('Precision', precision*100, '%')
    print('recall: ', recall*100, '%')
```

```python
def precision_recall_curve_plot(y_val, y_pred_proba):
    #thresold ndarray와 이 threshold에 따른 정밀도, 재현율 ndarray 추출.
    precisions, recalls, thresholds = precision_recall_curve(y_val, y_pred_proba)
    
    #x축을 threshold값으로, y축은 정밀도, 재현율 값으로 각각 plot 수행. 정밀도는 점선으로 표시
    plt.figure(figsize=(12,9))
    threshold_boundary =thresholds.shape[0]
    plt.plot(thresholds, precisions[0:threshold_boundary], linestyle='--', label='precision')
    plt.plot(thresholds, recalls[0:threshold_boundary], label='recall')
    
    #threshold 값 x축의 scale을 0.1단위로 변경
    start, end=plt.xlim()
    plt.xticks([0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95])
    
    # x축, y축 label과 legend, 그리고 grid 설정
    plt.xlabel('Threshold value');plt.ylabel('Precision and Recall value')
    plt.legend(); plt.grid()
    plt.show()
```
